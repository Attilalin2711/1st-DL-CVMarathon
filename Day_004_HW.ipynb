#### 資料的 row 數以及 column 數

```python
print('app_train has %s rows and %s cols'%(app_train.shape[0],app_train.shape[1]))
```

#### 列出所有欄位

```python
app_train.columns.map(lambda x:print(x))
```

Note
pandas 具有選項配置可更改dataframe的顯示設置,其中
1.display.max_columns與display.max_rows分別控制cols與rows的顯示數,預設值分別為60與20
2.透過pd.set_option能夠重設

```python
pd.set_option("display.max_columns",None)
app_train
```

#### 截取部分資料

```python
app_train.iloc[:,2]
```

```python
app_train.loc[:,'NAME_CONTRACT_TYPE']
```
